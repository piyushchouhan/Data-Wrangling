```python
import pandas as pd
```

```python
wine_data = pd.read_csv('wine.csv')
```

```python
wine_data.head()
```

## Simple fearure Scaling

```python
wine_data['fixed acidity']= wine_data['fixed acidity']/wine_data['fixed acidity'].max()
```

```python
wine_data.head()
```

```python
#entire data set
wine_data = wine_data / wine_data.max()
wine_data.head()
```

#Min max Scaling


```python
wine_data['fixed acidity'] = (wine_data['fixed acidity']-wine_data['fixed acidity'].min())/(wine_data['fixed acidity'].max()-wine_data['fixed acidity'].min())
```

```python
wine_data.head()
```

```python
#using sklearn

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

wine_data[wine_data.columns] = scaler.fit_transform(wine_data[wine_data.columns])
```

```python
wine_data.head()
```

## Z-Score Normalization


```python
 #Single feature manual
 wine_data['fixed acidity'] = (wine_data['fixed acidity']-wine_data['fixed acidity'].mean())\
                              /(wine_data['fixed acidity'].std())
```

```python
wine_data.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler  = StandardScaler()
wine_data[wine_data.columns] = scaler.fit_transform(wine_data[wine_data.columns])
```
